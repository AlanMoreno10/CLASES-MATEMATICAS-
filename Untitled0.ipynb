### 🧮 **Resolución Paso a Paso de $F(x) = x^4 - 4x^3 + 4x^2 + 4$**



#### 1️⃣  **Definición de la Función**
Definimos la función $F(x)$ como:
$$F(x) = x^4 - 4x^3 + 4x^2 + 4$$

```python
python
import sympy as sp

# Definimos la variable x
x = sp.Symbol('x')

# Definimos la función F(x)
F = x**4 - 4*x**3 + 4*x**2 + 4

# Mostramos la función
F
```

#### 2️⃣ **Derivada de la Función**
Calculamos la primera derivada de la función $F(x)$.
$$F'(x) = \frac{d}{dx} \left( x^4 - 4x^3 + 4x^2 + 4 \right)$$


```python
python
# Calculamos la primera derivada de F(x)
F_prime = sp.diff(F, x)

# Mostramos la derivada
F_prime

```

#### 3️⃣ **Segunda Derivada de la Función**
Calculamos la segunda derivada de la función $F(x)$.
$$F''(x) = \frac{d^2}{dx^2} \left( x^4 - 4x^3 + 4x^2 + 4 \right)$$



```python
python
# Calculamos la segunda derivada de F(x)
F_double_prime = sp.diff(F_prime, x)

# Mostramos la segunda derivada
F_double_prime

```

#### 4️⃣ **Puntos Críticos**
Encontramos los puntos críticos resolviendo $F'(x) = 0$.
$$F'(x) = 0$$


```python
python
# Resolvemos F'(x) = 0 para encontrar los puntos críticos
critical_points = sp.solve(F_prime, x)

# Mostramos los puntos críticos
critical_points
```

#### 5️⃣ **Evaluación de Puntos Críticos en la Segunda Derivada**
Evaluamos los puntos críticos en la segunda derivada para determinar la concavidad y clasificar los puntos críticos.


```python
python
# Evaluamos los puntos críticos en la segunda derivada
second_derivative_at_critical_points = [F_double_prime.subs(x, point) for point in critical_points]

```

#### 6️⃣ **Evaluación de la Función en los Puntos Críticos**
Evaluamos la función original $F(x)$ en los puntos críticos para obtener los valores correspondientes.

```python
python
# Evaluamos la función F(x) en los puntos críticos
function_values_at_critical_points = [F.subs(x, point) for point in critical_points]

# Mostramos la evaluación de la función en los puntos críticos
function_values_at_critical_points
```

# 📈 **Optimización: Una Variedad Especial de Análisis de Equilibrio**

## 🧠 **Definición de Optimización**
La **optimización** es un proceso matemático para encontrar el mejor resultado (máximo o mínimo) bajo ciertas restricciones. En economía, se utiliza para determinar la asignación óptima de recursos.

### ✏️ **Contexto Histórico**
La optimización tiene raíces antiguas, pero se formalizó como disciplina matemática en el siglo XX. Figuras clave incluyen a **Leonid Kantorovich**, quien desarrolló la programación lineal, y **John von Neumann**, quien trabajó en la teoría de juegos y optimización.

## 🧮 **Tipos de Optimización**

### ➕ **Optimización Sin Restricciones**
Este tipo de optimización no tiene limitaciones impuestas a las variables de decisión. La función objetivo se optimiza directamente.

### ➗ **Optimización con Restricciones**
En este caso, se busca optimizar una función objetivo sujeta a una serie de restricciones, que pueden ser de igualdad o desigualdad.

## 🔍 **Análisis de Equilibrio**
El análisis de equilibrio en economía examina cómo las fuerzas del mercado (oferta y demanda) interactúan para determinar los precios y cantidades de equilibrio. La optimización se utiliza para encontrar estos puntos de equilibrio.

### 🏛️ **Contexto Histórico del Equilibrio**
El concepto de equilibrio se remonta a **Adam Smith** y su "mano invisible". Posteriormente, **Alfred Marshall** y **Leon Walras** desarrollaron teorías más formales del equilibrio en mercados.

## 📉 **Función de Utilidad y Optimización**

### 🛒 **Ejemplo 1: Maximización de la Utilidad del Consumidor**
El objetivo del consumidor es maximizar su utilidad bajo una restricción presupuestaria. La función de utilidad $U(x, y)$ mide la satisfacción del consumidor, y el problema se puede plantear como:
$$ \max U(x, y) \text{ sujeto a } p_x x + p_y y \leq M $$
donde $p_x$ y $p_y$ son los precios de los bienes $x$ y $y$, y $M$ es el ingreso del consumidor.

```python
import sympy as sp

# Definimos las variables y parámetros
x, y, p_x, p_y, M = sp.symbols('x y p_x p_y M')
U = sp.Function('U')(x, y)

# Restricción presupuestaria
constraint = sp.Eq(p_x * x + p_y * y, M)

# Derivadas parciales
U_x = sp.diff(U, x)
U_y = sp.diff(U, y)

# Mostramos las derivadas parciales
U_x, U_y
```

### 📈 **Ejemplo Completo: Optimización con Restricciones en Economía**

Consideremos un problema de maximización de utilidad con una restricción presupuestaria. La función de utilidad es:
$$ U(x, y) = x^{0.5} y^{0.5} $$
y la restricción presupuestaria es:
$$ p_x x + p_y y = M $$

```python
# Definimos la función de utilidad específica
U = x**0.5 * y**0.5

# Restricción presupuestaria
constraint = sp.Eq(p_x * x + p_y * y, M)

# Creamos una variable simbólica para el multiplicador de Lagrange
lambda_ = sp.Symbol('lambda')

# Lagrangiana
L = U + lambda_ * (M - p_x * x - p_y * y)

# Derivadas parciales de la Lagrangiana
L_x = sp.diff(L, x)
L_y = sp.diff(L, y)
L_lambda = sp.diff(L, lambda_)

# Sistema de ecuaciones
system = [L_x, L_y, L_lambda]

# Resolución del sistema
solution = sp.solve(system, (x, y, lambda_))

# Mostramos la solución
solution
```
