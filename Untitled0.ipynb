###  **Ejemplo Completo: Optimizaci贸n con Restricciones en Econom铆a**

Consideremos un problema de maximizaci贸n de utilidad con una restricci贸n presupuestaria. La funci贸n de utilidad es:
$$ U(x, y) = x^{0.5} y^{0.5} $$

```python
# Definimos la funci贸n de utilidad espec铆fica
U = x**0.5 * y**0.5

# Restricci贸n presupuestaria
constraint = sp.Eq(p_x * x + p_y * y, M)

# Creamos una variable simb贸lica para el multiplicador de Lagrange
lambda_ = sp.Symbol('lambda')

# Lagrangiana
L = U + lambda_ * (M - p_x * x - p_y * y)

# Derivadas parciales de la Lagrangiana
L_x = sp.diff(L, x)
L_y = sp.diff(L, y)
L_lambda = sp.diff(L, lambda_)

# Sistema de ecuaciones
system = [L_x, L_y, L_lambda]

# Resoluci贸n del sistema
solution = sp.solve(system, (x, y, lambda_))

# Mostramos la soluci贸n
solution
```
